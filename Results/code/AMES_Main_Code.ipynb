## Modelling Preparation, Validation & Selection

<font color=blue>Prior to running predictions on the test set, the current training set will be used to derive the best predictor model that will be used as the standard linear regression model will not be accurate enough to identify the best fit features. The training data will be used as a cross-validation set in a train-test split to simulate the baseline model. This will be repeated across other regularization models for improvement.
</font>

```python
# Define matrix (X) and target vector (y)

features = [col for col in df._get_numeric_data().columns if col != 'saleprice']
X = df[features]
y = df['saleprice']
```

### Model Prep: Train/test split

```python
# Create the train test ratio split on training data

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Standard Scaling

```python
# Perform the scaling of the multiple features

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(X_train) # Extract the mean & std
X_train = ss.transform(X_train) # Calculate the Z-score from the fitted set
X_test = ss.transform(X_test)
```

<font color=blue>Here the feature is appropriately scaled into their z-scores for smoother input into the model predictors. The scaler also ensures reproducibility across the different type of predictors. </font>

### Build Linear Regression Baseline Model 

```python
# Import the regression model libraries and run linear regression cross validation

from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from sklearn.model_selection import cross_val_score

lr_model = LinearRegression()

lr_cv_mean_mse = -cross_val_score(lr_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean()
lr_cv_mean_rmse = lr_cv_mean_mse**0.5
lr_cv_mean_r2 = cross_val_score(lr_model, X_train, y_train, cv=5, scoring='r2').mean()

print("LR MSE: {}".format(lr_cv_mean_mse))
print("LR RMSE: {}".format(lr_cv_mean_rmse))
print("LR R2: {}".format(lr_cv_mean_r2))
```

### Inference of Linear Regression

<font color=blue>The negative R2 score of <b>-2e^20</b> is massively overblown. However this phenomenon is expected and due to overfitting associated with the linear regression model. This arises from having too many parameters and regression begins to model the random error (noise) in the data. Therefore, subsequent predictor models such as RidgeCV,LassoCV and ElasticNet will be used in order to improve on this baseline sore. </font>

### Run Ridge Regression Model

```python
# Generate the optimal alpha value

%%time
r_alphas = np.logspace(0, 5, 200) # set range of alphas for RidgeCV
ridge_model = RidgeCV(alphas=r_alphas, cv=5) #Initialize ridge model with alphas
ridge_model = ridge_model.fit(X_train, y_train) # run RidgeCV with test data  

ridge_optimal_alpha = ridge_model.alpha_  #extract the alpha

# Optimal value of alpha after the fitting
print ("Optimal Ridge Alpha: {}.".format(ridge_optimal_alpha))  
```

<font color=blue>An optimal alpha value of 153 is being determined for use in the subsequent cross-validation. Ridge assigns a penalty that is the squared magnitude of the coefficients to the loss function multiplied by lambda. As Lasso does, ridge also adds a penalty to coefficients the model overemphasizes. The value of lambda also plays a key role in how much weight you assign to the penalty for the coefficients. The larger your value of lambda, the more likely your coefficients get closer and closer to zero. 

<font color=red><b>Please note that alpha value will be different on each new run!</b></font>
    
</font>

### Cross-validate the Ridge regression value with Baseline

```python
# Cross validate Ridge with baseline

ridge_opt = Ridge(alpha=ridge_model.alpha_)

ridge_cv_mean_mse = -cross_val_score(ridge_opt, X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean()
ridge_cv_mean_r2 = cross_val_score(ridge_opt, X_train, y_train, cv=5, scoring='r2').mean()
ridge_cv_mean_rmse = (ridge_cv_mean_mse)**0.5


print("Ridge MSE: {}".format(ridge_cv_mean_mse))
print("Ridge RMSE: {}".format(ridge_cv_mean_rmse))
print("Ridge R2: {}".format(ridge_cv_mean_r2))
```

### Inference of Ridge Regression

<font color=blue>The R2 score has significantly improved to 0.903. We can observe that with the use of a penalty parameter, the   ridge model performs well compared to the ordinary least square method. Ridge estimator is especially good at improving the least-squares estimate when multicollinearity is present.

<font color=red><b>Please note that R2 values will marginally change on each new run!</b></font>

</font>

### Run Lasso Regression Model

```python
# Generate the alpha value for Lasso

lasso_model = LassoCV(n_alphas=500, cv=5)
lasso_model = lasso_model.fit(X_train, y_train)

lasso_optimal_alpha = lasso_model.alpha_

print ("Optimal Lasso alpha: {}.".format(lasso_optimal_alpha))  
```

<font color=blue> Lasso assigns a penalty to the coefficients in the linear model and eliminates variables with coefficients that zero. Lasso adds a penalty to coefficients the model overemphasizes. This reduces the degree of overfitting that occurs within the model. </font>

### Cross-validate the Lasso regression value with Baseline

```python
# Cross validate Lasso with baseline

lasso_model = Lasso(alpha=lasso_optimal_alpha)

lasso_mean_mse = -cross_val_score(lasso_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean()
lasso_mean_rmse = lasso_mean_mse**0.5
lasso_mean_r2 = cross_val_score(lasso_model, X_train, y_train, cv=5, scoring='r2').mean()

print("Lasso MSE: {}".format(lasso_mean_mse))
print("Lasso RMSE: {}".format(lasso_mean_rmse))
print("Lasso R2: {}".format(lasso_mean_r2))  
```

### Inference of Lasso Regression

<font color=blue>The R2 score of 0.906 performs marginally better than the ridge regression method. The Lasso model can also provide good prediction with small margin of error, because shrinking and removing the coefficients can reduce variance without a substantial increase of the bias. This increases the model interpretability by eliminating irrelevant variables that are not associated with the response variable since we also have many variables such as dummy categorical data.

<font color=red><b>Please note that R2 values will marginally change on each new run!</b></font>

</font>

### Run Elastic Net Regression Model

```python
# Initialize the Enet model parameters

enet_ratio = np.linspace(0.01, 1.0, 25)
enet_model = ElasticNetCV(n_alphas=100, l1_ratio=enet_ratio, cv=5)
enet_model = enet_model.fit(X_train, y_train)

enet_optimal_alpha = enet_model.alpha_
enet_optimal_ratio = enet_model.l1_ratio_

print ("Optimal Enet Alpha: {}.".format(enet_optimal_alpha))  
print ("Optimal Enet Ratio: {}.".format(enet_optimal_ratio))  
```

<font color=blue>The Elastic Net combines characteristics of both lasso and ridge and works well to eliminate collinearity. It also reduces the impact of different features while not eliminating all of the features.</font>

### Cross-validate the Elastic Net regression value with Baseline

```python
# Cross validate Elastic Net with baseline

enet = ElasticNet(alpha=enet_optimal_alpha, l1_ratio=enet_optimal_ratio)

enet_mean_mse = -cross_val_score(enet, X_train, y_train, cv=5, scoring='neg_mean_squared_error').mean()
enet_mean_rmse = enet_mean_mse**0.5
enet_mean_r2 = cross_val_score(enet, X_train, y_train, cv=5, scoring='r2').mean()

print("Enet MSE: {}".format(enet_mean_mse))
print("Enet RMSE: {}".format(enet_mean_rmse))
print("Enet R2: {}".format(enet_mean_r2))
```

### Inference of Elastic Net Regression

<font color=blue>The Elastic Net method yields an almost identical value to the Lasso Model but narrowly loses out by 5 decimal places.

<font color=red><b>Please note that R2 values will marginally change on each new run!</b></font>
</font>
